```python
from langchain_huggingface import HuggingFaceEmbeddings
from dotenv import load_dotenv
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
load_dotenv()
```

```python
embedding = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
documents = [
    "Virat Kohli is one of the most successful and consistent batsmen in modern cricket.",
    "He made his international debut for India in 2008.",
    "Known for his aggressive batting style, Kohli has scored over 70 international centuries.",
    "He served as the captain of the Indian cricket team across all formats until 2022.",
    "Off the field, Kohli is also known for his fitness, philanthropy, and brand endorsements."
]
```

```python
query = "Tell me about virat kholi"
```

```python
doc_embedding = embedding.embed_documents(documents)
```

```python
doc_embedding
```

```python
query_embedding = embedding.embed_query(query)
```

```python
scores = cosine_similarity([query_embedding], doc_embedding)[0]
```

```python
index, score = sorted(list(enumerate(scores)), key = lambda x : x[1])[-1]
```

```python
print(documents[index])
print(f"Similarity Score is :{score*100 : .2f}%")
```
